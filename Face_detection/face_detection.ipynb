```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
face_cascade=cv.CascadeClassifier('D:\AI\OpenCV_Projects\Face_detection\haarcascade_frontalface_default.xml')
```

```python
img=cv.imread('D:\AI\OpenCV_Projects\Face_detection\9hsbjttgnuoa1nj0_1607070497.webp')
```

```python
plt.imshow(img)
```

```python
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
faces=face_cascade.detectMultiScale(img,1.1,4,minSize=(30,30))
```

```python
len(faces)
```

```python
for (x,y,w,h) in faces:
    cv.rectangle(img,(x-5,y-5),(x+w+5,y+h+5),(0,255,0),3)
plt.imshow(img)
```
