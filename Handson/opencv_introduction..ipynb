```python
# loading libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the image
img=cv.imread('D:\AI\OpenCV_Projects\Handson\mohanlal.jpg')
```

```python
# display image
cv.imshow("image",img)
cv.waitKey()
cv.destroyAllWindows()
```

Drawing Rectangle in the image

```python
img_rect=cv.rectangle(img,pt1=(50,250),pt2=(150,150),color=(0,0,255),thickness=3)
plt.imshow(img_rect)
```

Drawing Circle

```python
img_cir=cv.circle(img,center=(150,150),radius=50,thickness=3,color=(0,255,0))
plt.imshow(img_cir)
```

Adding text

```python
img_text=cv.putText(img,text="Mohanlal",org=(50,255),fontFace=cv.FONT_HERSHEY_SCRIPT_SIMPLEX,fontScale=1,color=(255,0,0),thickness=2)
plt.imshow(img_text)
```

```python
# Convert the images to gray scale
gray=cv.imread('D:\AI\OpenCV_Projects\Handson\mohanlal.jpg',cv.IMREAD_GRAYSCALE)
cv.imshow("image",gray)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# Convert the images to black n white
(thresh,bw)=cv.threshold(gray,128,255,cv.THRESH_BINARY|cv.THRESH_OTSU)
cv.imshow("image",bw)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# Edge detection
edges=cv.Canny(gray,150,150)
cv.imshow("image",edges)

cv.waitKey()
cv.destroyAllWindows()
```
